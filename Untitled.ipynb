```python
import pandas as pd
import numpy as np


df = pd.read_csv("advanced_b2b_grocery_sku_dataset (1) (1).csv")


df.head()
```

```python
df.columns
```

```python
df[['Length_m', 'Width_m', 'Height_m', 'Cube_m3']].head()
```

```python

storage_ranges = df.groupby('Preferred_Storage_Type')['Cube_m3'].agg(['min','max','mean'])
print(storage_ranges)
```

```python
def classify_storage(cube):
    if cube >= 0.05:      
        return 'Pallet'
    elif cube >= 0.02:
        return 'Heavy Rack'
    else:
        return 'Regular Rack'

df['Storage_Model'] = df['Cube_m3'].apply(classify_storage)
```

```python
from sklearn.linear_model import LinearRegression
```

```python


df['Effective_Demand'] = df['Base_Daily_Demand'] * df['Promo_Spike_Multiplier']


df_encoded = pd.get_dummies(df, columns=['Weekly_Seasonality_Type'], drop_first=True)


X = df_encoded[
    [
        'Base_Daily_Demand',
        'Promo_Spike_Multiplier',
        'Weekly_Seasonality_Type_Weekday Peak',
        'Weekly_Seasonality_Type_Weekend Peak'
    ]
]


y = df_encoded['Effective_Demand']


model = LinearRegression()
model.fit(X, y)


df_encoded['Predicted_Demand'] = model.predict(X)


df_encoded[['SKU_ID', 'Predicted_Demand']].head()
```

```python
df['Demand_Volatility']
```

```python
df['Demand_std'] = df['Demand_Volatility']
```

```python
df['Predicted_Demand'] = df_encoded['Predicted_Demand']
```

```python
Z = 1.65  # 95% service level

df['Safety_Stock'] = Z * np.sqrt(
    (df['Vendor_TAT_mean_days'] * (df['Demand_std'] ** 2)) +
    ((df['Predicted_Demand'] ** 2) * (df['Vendor_TAT_variability_days'] ** 2))
)


```

```python
df['ROP'] = (
    df['Predicted_Demand'] * df['Vendor_TAT_mean_days']
) + df['Safety_Stock']
```

```python
df['Reorder_Quantity'] = np.maximum(df['ROP'], df['Vendor_MOQ_units'])
```

```python
df['DOI'] = df['Reorder_Quantity'] / df['Predicted_Demand']
```

```python
def risk_flag(row):

    if row['Vendor_MOQ_units'] > row['ROP']:
        return 'MOQ Risk'
    
    elif row['Vendor_TAT_mean_days'] > row['DOI']:
        return 'Lead Time Risk'
    
    elif row['Cube_m3'] > df['Cube_m3'].quantile(0.9):
        return 'Space Risk'
    
    elif row['Shelf_Life_days'] < row['DOI']:
        return 'Expiry Risk'
    
    else:
        return 'Low Risk'

df['Risk_Flag'] = df.apply(risk_flag, axis=1)
```

```python
df['Total_Volume_Required'] = df['Reorder_Quantity'] * df['Cube_m3']

total_volume = df['Total_Volume_Required'].sum()

print("Total Volume Required:", total_volume)
```

```python
final_output = df[[
    'SKU_ID',
    'Predicted_Demand',
    'Safety_Stock',
    'ROP',
    'Reorder_Quantity',
    'DOI',
    'Preferred_Storage_Type',
    'Risk_Flag'
]]

final_output.head()
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML
import numpy as np

def interactive_ui(df):
    
    display(HTML("<h3>📊 AI Inventory Planning Tool</h3>"))
    
    demand_slider = widgets.FloatSlider(value=0, min=-50, max=100, step=5, description='Demand %')
    tat_slider = widgets.FloatSlider(value=0, min=-5, max=10, step=1, description='TAT Change')
    moq_slider = widgets.FloatSlider(value=0, min=-50, max=100, step=5, description='MOQ %')
    promo_slider = widgets.FloatSlider(value=0, min=-50, max=100, step=5, description='Promo %')
    
    button = widgets.Button(description="Run Simulation", button_style='success')
    
    output = widgets.Output()
    
    def on_button_click(b):
        with output:
            output.clear_output()
            
            df_copy = df.copy()
            
            
            df_copy['Base_Daily_Demand'] *= (1 + demand_slider.value / 100)
            df_copy['Vendor_TAT_mean_days'] += tat_slider.value
            df_copy['Vendor_MOQ_units'] *= (1 + moq_slider.value / 100)
            df_copy['Promo_Spike_Multiplier'] *= (1 + promo_slider.value / 100)
            
            
            df_copy['Predicted_Demand'] = (
                df_copy['Base_Daily_Demand'] *
                df_copy['Promo_Spike_Multiplier']
            )
            
            
            df_copy['Demand_std'] = df_copy['Demand_Volatility']
            
            # Safety Stock
            Z = 1.65
            df_copy['Safety_Stock'] = Z * np.sqrt(
                (df_copy['Vendor_TAT_mean_days'] * (df_copy['Demand_std'] ** 2)) +
                ((df_copy['Predicted_Demand'] ** 2) * (df_copy['Vendor_TAT_variability_days'] ** 2))
            )
            
            # ROP
            df_copy['ROP'] = (
                df_copy['Predicted_Demand'] * df_copy['Vendor_TAT_mean_days']
            ) + df_copy['Safety_Stock']
            
            # Reorder
            df_copy['Reorder_Quantity'] = np.maximum(
                df_copy['ROP'], df_copy['Vendor_MOQ_units']
            )
            
            # DOI
            df_copy['DOI'] = df_copy['Reorder_Quantity'] / df_copy['Predicted_Demand']
            
            print("✅ Simulation Complete\n")
            display(df_copy[['SKU_ID','Predicted_Demand','ROP','Reorder_Quantity','DOI']].head())
    
    button.on_click(on_button_click)
    
    display(demand_slider, tat_slider, moq_slider, promo_slider, button, output)
```

```python
interactive_ui(df)
```
